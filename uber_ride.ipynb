```python
q_low = df['fare_amount'].quantile(0.01)
q_high = df['fare_amount'].quantile(0.99)
df = df[(df['fare_amount'] < q_high) & (df['fare_amount'] > q_low)]
```

```python
df
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
 #Take x as predictor variable
 x = df.drop("fare_amount", axis = 1)
 #And y as target variable
 y = df['fare_amount']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 1)
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lrmodel = LinearRegression()
lrmodel.fit(x_train,y_train)
```

```python
predict = lrmodel.predict(x_test)
```

```python
print(predict)
```

```python
from sklearn.metrics import mean_squared_error
lrmodelrmse = np.sqrt(mean_squared_error(predict,y_test))
print("RMSE error for the model is ",lrmodelrmse)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfrmodel = RandomForestRegressor(n_estimators = 100 , random_state = 101)
```

```python
rfrmodel.fit(x_train,y_train)
predict = rfrmodel.predict(x_test)
```

```python
rfrmodel_rmse = np.sqrt(mean_squared_error(predict,y_test))
print("RMSE value for random forest model is:",rfrmodel_rmse)
```

```python
from sklearn.metrics import r2_score
lrmodelr2 = r2_score(y_test,predict)
```

```python
print(lrmodelr2)
```

```python
rfrmodelr2 = r2_score(y_test,predict)
```
